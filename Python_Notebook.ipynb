```python
DOSYA_ADI = "gorevler.txt"

def gorevleri_yukle():
    gorevler = []
    try:
        with open(DOSYA_ADI, "r", encoding="utf-8") as dosya:
            for satir in dosya:
                metin, durum = satir.strip().rsplit(" | ", 1)
                gorevler.append((metin, durum == "True"))
    except FileNotFoundError:
        print("Görev dosyası bulunamadı. Yeni bir liste oluşturulacak.")
    return gorevler
```

```python
def gorevleri_listele(gorevler):
    if not gorevler:
        print("Henüz görev bulunmamaktadır.")
        return
    print("\n--- GÖREV LİSTESİ ---")
    for i, (metin, durum) in enumerate(gorevler, 1):
        simge = "✅" if durum else "❌"
        print(f"{i}. {simge} {metin}")
    print("---------------------")
```

```python
def gorev_ekle(gorevler):
    yeni = input("Yeni görev: ").strip()
    if not yeni:
        print("Boş görev eklenemez.")
        return
    gorevler.append((yeni, False))
    gorevleri_kaydet(gorevler)
    print(f"'{yeni}' görevi eklendi.")
```

```python
def gorev_duzenle(gorevler):
    gorevleri_listele(gorevler)
    try:
        no = int(input("Düzenlemek istediğiniz görevin numarası: "))
        if 1 <= no <= len(gorevler):
            yeni = input("Yeni görev metni: ").strip()
            if not yeni:
                print("Boş görev eklenemez.")
                return
            _, durum = gorevler[no - 1]
            gorevler[no - 1] = (yeni, durum)
            gorevleri_kaydet(gorevler)
            print("Görev güncellendi.")
        else:
            print("Geçersiz görev numarası!")
    except ValueError:
        print("Lütfen geçerli bir sayı girin.")
```

```python
def gorev_sil(gorevler):
    gorevleri_listele(gorevler)
    try:
        no = int(input("Silmek istediğiniz görevin numarası: "))
        if 1 <= no <= len(gorevler):
            silinen, _ = gorevler.pop(no - 1)
            gorevleri_kaydet(gorevler)
            print(f"'{silinen}' görevi silindi.")
        else:
            print("Geçersiz görev numarası!")
    except ValueError:
        print("Lütfen geçerli bir sayı girin.")
```

```python
def gorev_tamamla(gorevler):
    gorevleri_listele(gorevler)
    try:
        no = int(input("Durumunu değiştirmek istediğiniz görev numarası: "))
        if 1 <= no <= len(gorevler):
            metin, durum = gorevler[no - 1]
            gorevler[no - 1] = (metin, not durum)
            gorevleri_kaydet(gorevler)
            print(f"'{metin}' durumu değiştirildi.")
        else:
            print("Geçersiz görev numarası!")
    except ValueError:
        print("Lütfen geçerli bir sayı girin.")
```

```python
def ana_menu():
    gorevler = gorevleri_yukle()
    while True:
        print("\n--- TO-DO LİST UYGULAMASI ---")
        print("1. Görevleri Listele")
        print("2. Yeni Görev Ekle")
        print("3. Görev Düzenle")
        print("4. Görev Sil")
        print("5. Görev Durumunu Değiştir (✅/❌)")
        print("6. Çıkış")
        secim = input("Seçiminiz (1-6): ").strip()

        if secim == "1":
            gorevleri_listele(gorevler)
        elif secim == "2":
            gorev_ekle(gorevler)
        elif secim == "3":
            gorev_duzenle(gorevler)
        elif secim == "4":
            gorev_sil(gorevler)
        elif secim == "5":
            gorev_tamamla(gorevler)
        elif secim == "6":
            print("Programdan çıkılıyor...")
            break
        else:
          print("Geçersiz seçim! Lütfen 1-6 arasında bir sayı girin.")
```

```python
if __name__ == "_main_":
    ana_menu()
```
